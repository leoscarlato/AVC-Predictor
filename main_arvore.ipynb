```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df = df.dropna()
```

```python
df_clean = pd.get_dummies(df).astype(float)
df_clean = df_clean.drop(columns=['id','age','avg_glucose_level', 'bmi'])
df_clean
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

ExplicaÃ§Ã£o cÃ©lula 1:

ImportÃ¢ncia: Gerar as features. Para isso iremos remover a coluna stroke do nosso dataset, df_clean pois ela Ã© o target, e nÃ£o uma feature.

Resultado: pandas DataFrame com as features

--> cÃ©lula 1

```python
# dataset 
# data, all columns except stroke
X = df_clean.drop(columns=["stroke"])
type(X)
```

ExplicaÃ§Ã£o cÃ©lula 2:

ImportÃ¢ncia: Gerar o target. Para isso iremos selecionar apenas a coluna stroke do nosso dataset, df_clean pois ela Ã© o target, e nÃ£o uma feature.

Resultado: pandas Series com o target

--> cÃ©lula 2

```python
# target, only stroke column
Y = df_clean["stroke"]
Y
```

ExplicaÃ§Ã£o cÃ©lula 3:

ImportÃ¢ncia: Para que o nosso modelo nÃ£o seja treinado com os prÃ³prios dados que serÃ£o usados para avaliar sua performance, Ã© necessÃ¡rio separar os dados em treino e teste. Para isso iremos utilizar a funÃ§Ã£o train_test_split, que irÃ¡ separar os dados em treino e teste, de acordo com os parÃ¢metros passados.

Resultado: 

X_train, (features de treino)

X_test, (features de teste)

y_train, (target de treino)

y_test, (target de teste)



--> cÃ©lula 3

```python
# split dataset into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.5)
```

ExplicaÃ§Ã£o cÃ©lula 4:

ImportÃ¢ncia: Treinar o modelo. Para isso iremos utilizar a funÃ§Ã£o fit, que irÃ¡ treinar o modelo com os dados de treino. Iremos inicializar o modelo DecisionTreeClassifier, e passar os dados de treino para a funÃ§Ã£o fit.

Resultado: Modelo treinado

--> cÃ©lula 4

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)
```

ExplicaÃ§Ã£o cÃ©lula 5:

ImportÃ¢ncia: Avaliar o modelo. Para isso iremos utilizar a funÃ§Ã£o score, que irÃ¡ avaliar o modelo com os dados de teste. Iremos passar os dados de teste para a funÃ§Ã£o score.

Resultado: Score do modelo

--> cÃ©lula 5

```python
print(f"Accuracy: {clf.score(X_test, Y_test)}")
```

ExplicaÃ§Ã£o cÃ©lula 6:

ImportÃ¢ncia: Avaliar as features mais importantes para o modelo. Para isso iremos utilizar o atributo feature_importances_, que irÃ¡ avaliar as features mais importantes para o modelo.


Resultado: Lista com as features mais importantes para o modelo e o seu respectivo valor de importÃ¢ncia. Ordenado de forma decrescente.

--> cÃ©lula 6

```python
importance = []
for i in range(len(clf.feature_importances_)):
   importance.append((X.columns[i],clf.feature_importances_[i]))

# display the importance of each attribute, sorted 
importance.sort(key=lambda x: x[1], reverse=True)
importance
```
