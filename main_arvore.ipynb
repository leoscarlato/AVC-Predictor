```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
df = df.dropna()
```

```python
df_clean = pd.get_dummies(df).astype(float)
df_clean = df_clean.drop(columns=['id','age','avg_glucose_level', 'bmi'])
df_clean
```

# Modelo de classificação DecisionTreeClassifier

Explicação célula 0:

Importância: 

from sklearn.tree import DecisionTreeClassifier: importa o modelo DecisionTreeClassifier, que é um modelo de classificação que será utilizado.

from sklearn.model_selection import train_test_split: importa a função train_test_split, que será utilizada para separar os dados em treino e teste.

Resultado:  Módulos necessários para o projeto importados .

--> célula 0

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

Explicação célula 1:

Importância: Gerar as features. Para isso iremos remover a coluna stroke do nosso dataset, df_clean pois ela é o target, e não uma feature.

Resultado: pandas DataFrame com as features

--> célula 1

```python
# dataset 
# data, all columns except stroke
X = df_clean.drop(columns=["stroke"])
type(X)
```

Explicação célula 2:

Importância: Gerar o target. Para isso iremos selecionar apenas a coluna stroke do nosso dataset, df_clean pois ela é o target, e não uma feature.

Resultado: pandas Series com o target

--> célula 2

```python
# target, only stroke column
Y = df_clean["stroke"]
Y
```

Explicação célula 3:

Importância: Para que o nosso modelo não seja treinado com os próprios dados que serão usados para avaliar sua performance, é necessário separar os dados em treino e teste. Para isso iremos utilizar a função train_test_split, que irá separar os dados em treino e teste, de acordo com os parâmetros passados.

Resultado: 

X_train, (features de treino)

X_test, (features de teste)

y_train, (target de treino)

y_test, (target de teste)



--> célula 3

```python
# split dataset into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.5)
```

Explicação célula 4:

Importância: Treinar o modelo. Para isso iremos utilizar a função fit, que irá treinar o modelo com os dados de treino. Iremos inicializar o modelo DecisionTreeClassifier, e passar os dados de treino para a função fit.

Resultado: Modelo treinado

--> célula 4

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)
```

Explicação célula 5:

Importância: Avaliar o modelo. Para isso iremos utilizar a função score, que irá avaliar o modelo com os dados de teste. Iremos passar os dados de teste para a função score.

Resultado: Score do modelo

--> célula 5

```python
print(f"Accuracy: {clf.score(X_test, Y_test)}")
```

Explicação célula 6:

Importância: Avaliar as features mais importantes para o modelo. Para isso iremos utilizar o atributo feature_importances_, que irá avaliar as features mais importantes para o modelo.


Resultado: Lista com as features mais importantes para o modelo e o seu respectivo valor de importância. Ordenado de forma decrescente.

--> célula 6

```python
importance = []
for i in range(len(clf.feature_importances_)):
   importance.append((X.columns[i],clf.feature_importances_[i]))

# display the importance of each attribute, sorted 
importance.sort(key=lambda x: x[1], reverse=True)
importance
```
